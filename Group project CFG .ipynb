```python
from google.colab import drive
drive.mount('/content/drive')

from google.colab import files
uploaded = files.upload()

import pandas as pd
sales_report=pd.read_csv('sales.csv')
sales_report.head()

# Reading the data using read_csv from pandas and showing a head (first 5 entries of the data)
```

```python
m_sales=[]
m_sales=sales_report["sales"]

# Creating an array called m_sales and filling to make a list with the 'sales' column from the sales.csv in the previous step
```

```python
print(m_sales)

# Printing the list
```

```python
tot_sales=sum(m_sales)
print("Total Sales=",tot_sales)
avg_sales=tot_sales/len(m_sales)
print("Average Sales=",round(avg_sales))

# Finding the summation of the sales and the average
```

```python
sales_report["Monthly % change"]=sales_report["sales"].pct_change()
print(sales_report)

# Creating a new column showing the monthly changes as a percentage
```

```python
max_sales=sales_report['sales'].max()
min_sales=sales_report['sales'].min()


print(max_sales, min_sales)

# Finding the highest and lowest sales
```

```python
sales_report["Revenue"]=sales_report["sales"]-sales_report["expenditure"]
```

```python
import matplotlib.pyplot as plt
sales_report.plot(x="month",y="sales")
plt.show()

# Visually displaying the sales by month - demonstrating that July had the
# highest sales and February had the lowest sales
```
